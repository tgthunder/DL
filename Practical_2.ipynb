# Converting categorical data to numerical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab = LabelEncoder()
```

```python
df['T'] = lab.fit_transform(df['T'])
```

```python
df
```

```python
x = df.drop(["T"], axis=1)
```

```python
y= df["T"]
```

```python
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
x_scale = sc.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x_scale, y,random_state=42,test_size=0.5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import tensorflow
from tensorflow.keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(80,activation='relu',input_dim=x_train.shape[1]))
model.add(Dense(40,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#training the model
model.fit(x_train, y_train, epochs = 100, batch_size = 32, validation_data =(x_test, y_test))
```

```python
#evaluate the model
loss = model.evaluate(x_test, y_test)
print(f"MSE of test data = {loss}")
```
